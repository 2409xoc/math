```python
import requests, time, wget, os
from bs4 import BeautifulSoup
```

```python
req = BeautifulSoup(requests.get("https://cemc.uwaterloo.ca/contests/past_contests.html").content, "html.parser")
tables = req.find_all("table", {"class": "vsmalltable"})[-1].find_all("a")[:27]
ll = []
directory = "/Users/henry/fun/math/ccc/pdfs/"
```

```python
for i in tables:
    if (i['href'].split("/")[1] == "contests"):
        href = "https://cemc.uwaterloo.ca" + i["href"]
    else:
        href = "https://cemc.uwaterloo.ca/contests/" + i["href"]
    req = BeautifulSoup(requests.get(href).content, "html.parser")
    links = (req.find_all("a"))
    for j in links:
        try:
            if ("ccc" in j['href'].split("/")) or ("stage 1" in j['href'].split("/")) or ("stage1" in j['href'].split("/")):
                if ("pdf" in j['href'].split(".")):
                    if ("marking" not in j['href'].split("/")[-1].lower()) and ("commentary" not in j['href'].split("_")[-1].lower()):
                        if (str(j['href'])[0] != "/"):
                            l = href.split("index.html")[0] + j['href']
                            l = l.replace(" ", "%20")
                            ll.append(l)
        except:
            pass
            
```

```python
for i in ll:
    path = directory + (i.split("/")[-3])
    if (os.path.isdir(path) == False):
        os.makedirs(path)
    try:
        os.system(f'wget {i} -P {path}')
    except:
        print(i)
```
